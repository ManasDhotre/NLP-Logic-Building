```python
import re

def preprocess_text(text):
    if not isinstance(text, str) or text.strip() =="":
        return [],""


    #Remove mails and links
    text = re.sub(r"https\S+|www\S+|http\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)

    #Remove number
    text = re.sub(r"\d+", "", text)

    #Lowercase
    text = text.lower()

    #Repeated 9Characters
    text = re.sub(r"(.)\1{2,}", r"\1\1", text)
    
    #Remove special character
    text = re.sub(r"[^a-z\s]", "", text)

    #Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    #Tokenization
    tokens = text.split()

    #Remove short tokens
    cleaned_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]

    cleaned_statement = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_statement
    
```

```python
test_statements = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in test_statements:
    tokens, clean = preprocess_text(sentence)
    print("Original : ", sentence)
    print("Tokens : ", tokens)
    print("Clean :", clean)
    print("-"*50)
    
```

```python
def token_analytics(tokens):
    if not tokens:
        return 0, 0, 0
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total
    return total, unique, round(avg_len, 2)

for sentence in test_statements:
    tokens, _ = preprocess_text(sentence)
    total, unique, avg = token_analytics(tokens)

    print(f"Sentence: {sentence}")
    print(f"Total: {total}, Unique: {unique}, Average Length: {avg}")
    print("-"*50)
```

```python
from collections import Counter

all_tokens = []

for sentence in test_statements:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 frequent words:")
print(counter.most_common(10))

print("\nLeast 5 frequent words:")
print(counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
